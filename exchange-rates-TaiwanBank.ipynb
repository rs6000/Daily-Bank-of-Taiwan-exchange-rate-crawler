```python
#抓取台灣銀行牌告匯率 by 大數學堂
#http://www.largitdata.com
#原版 20180520
#20251005 更新

import pandas as pd
from datetime import datetime
import sqlite3

#source
url="https://rate.bot.com.tw/xrt?Lang=zh-TW"

#讀取網頁
dfs=pd.read_html(url)

#測試
#type(dfs)
#len(dfs)

#資料呈現
#dfs[0]

currency=dfs[0]
```

```python
#檢查
#type(currency)

#擷取需要的欄位
currency_fix=currency.iloc[:,0:3]

#檢查
#currency_fix

#自訂欄位名稱
currency_fix.columns=[u'幣別',u'買入',u'賣出']

#檢查
#currency_fix

#清除幣別欄重複字元
#currency_fix[u'幣別']=currency_fix[u'幣別'].str.extract('\((\w+)\)')
# gemini 建議寫法，原本寫法輸出一樣。但會出現語法可能有誤的訊息
currency_fix[u'幣別'] = currency_fix[u'幣別'].str.extract(r'\((\w+)\)')

#檢查
currency_fix.head(15)
```

```python
# ZAR 跟 SEK 出現 空值 -
# 要出現空值的欄位過濾掉不要顯示在資料表裡
# 最後 .reset_index(drop=True) 是把索引數字重新排序
# 因為刪掉 8 ZAR ,9 SEK這兩個欄位後，原本索引的數字 8 9會直接跳過，索引要重新排序
# gemini 提供寫法
currency_final = currency_fix[(currency_fix[u'買入'] != '-') & (currency_fix[u'賣出'] != '-')].reset_index(drop=True)
```
